```
from pyspark.sql.functions import *
from pyspark.sql.types import TimestampType
```

```
payments = spark.table("default.staging_payments")
display(payments.limit(10))
```

```
# Check whether payment_id is duplicated
print(payments.count())
fact_payments = payments.dropDuplicates(["payment_id"])
print(fact_payments.count())
```

```
# Because the payment_id field is unique, so don't need to do anything for this table, we can save it as fact_payments table
fact_payments.write.format("delta").mode("overwrite").saveAsTable("default.fact_payments")
```

```
trips = spark.table("default.staging_trips")
display(trips.limit(10))
```

```
# Check whether trip_id is duplicated
print(trips.count())
fact_trips = trips.dropDuplicates(["trip_id"])
print(fact_trips.count())
```

```
# Because the station_id field is unique, so we can use this field for identity row
# I add 2 new fields: trip_duration and account_age to create fact_trips table
fact_trips = fact_trips.alias("tbl1").join(riders.alias("tbl2"), col("tbl1.rider_id") == col("tbl2.rider_id"), "left") \
            .withColumn('trip_duration_minute', round((unix_timestamp("ended_at") - unix_timestamp('start_at'))/60)) \
            .withColumn('account_age_day', round((unix_timestamp("account_start_date") - unix_timestamp('birthday'))/3600/24)) \
            .select("trip_id", col("tbl1.rider_id"), "rideable_type", "start_station_id", "end_station_id", "start_at", "ended_at", "trip_duration_minute", "account_age_day")
            
display(fact_trips.limit(10))
```

```
fact_trips.write.format("delta").mode("overwrite").saveAsTable("default.fact_trips")
```

```
# get all datetime of start_at field (trips table) amd date field (payments table) to create dim_datetimes table
trip_start_at = trips.select('start_at').rdd.flatMap(lambda x: x).collect()
payment_date = payments.select('date').rdd.flatMap(lambda x: x).collect()
date = trip_start_at + payment_date
time_df = spark.createDataFrame(date, TimestampType())
dim_datetimes = time_df.toDF("datetime") \
                .withColumn('hour', hour('datetime')) \
                .withColumn('day', dayofmonth('datetime')) \
                .withColumn('week', weekofyear('datetime')) \
                .withColumn('day_of_week', dayofweek('datetime')) \
                .withColumn('month', month('datetime')) \
                .withColumn('quarter', quarter('datetime')) \
                .withColumn('year', year('datetime')) \
                .distinct()
display(dim_datetimes.limit(10))
```

```
dim_datetimes.write.format("delta").mode("overwrite").saveAsTable("default.dim_datetimes")
```
